```python
import pandas as pd
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
url='https://www.flipkart.com/watches/wrist-watches/pr?sid=r18%2Cf13&p%5B%5D=facets.ideal_for%255B%255D%3DWomen&otracker=categorytree&fm=neo%2Fmerchandising&iid=M_a43f3d72-03cf-4b83-851e-f59dbb8ac411_1_372UD5BXDFYS_MC.STURTSNOXQD2&otracker=hp_rich_navigation_1_1.navigationCard.RICH_NAVIGATION_Fashion%7EWatches%2Band%2BAccessories%7EMen%2B%2526%2BWomen%2BWatches_STURTSNOXQD2&otracker1=hp_rich_navigation_PINNED_neo%2Fmerchandising_NA_NAV_EXPANDABLE_navigationCard_cc_1_L2_view-all&cid=STURTSNOXQD2'
```

```python
page=requests.get(url)
```

```python
page
```

```python
content=BeautifulSoup(page.content,'html.parser')
```

```python
content
```

```python
data =content.find_all('div',{'class':'_1xHGtK _373qXS'})
```

```python
data
```

```python
links=[]
watch_name=[]
start_link='https://www.flipkart.com'
```

```python
for items in data:
  rest_link = items.find('a')['href']
  name=items.find('div',attrs={'class':'_2WkVRV'})
  watch_name.append(name.text)
  links.append(start_link+rest_link)
```

```python
print(watch_name[0])
print(links[0])
```

```python
dataframe={"Watch Name": watch_name,'Links': links}
Final_dataframe= pd.DataFrame(dataframe)
print(Final_dataframe)
```

```python
Final_dataframe.to_csv('fk_data_url.csv')
```
